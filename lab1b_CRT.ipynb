```python
def mod_inverse(a, m):
    """Returns modular inverse of a under modulo m, or None if not exists."""
    t1, t2 = 0, 1
    r1, r2 = m, a
    while r2 != 0:
        q = r1 // r2
        r = r1 % r2
        t = t1 - q * t2
        r1, r2 = r2, r
        t1, t2 = t2, t
    return t1 % m if r1 == 1 else None

def crt_reconstruct(c_list, moduli):
    """Reconstructs final result from residues using Chinese Remainder Theorem."""
    M = 1
    for m in moduli:
        M *= m

    result = 0
    for i in range(len(moduli)):
        mi = moduli[i]
        Mi = M // mi
        Mi_inverse = mod_inverse(Mi, mi)
        if Mi_inverse is None:
            raise ValueError(f"Modular inverse for Mi={Mi} mod {mi} doesn't exist.")
        result += c_list[i] * Mi * Mi_inverse

    return result % M

def to_residue_repr(number, moduli):
    """Convert a number into its residue representation (a1, a2, ..., ak)"""
    return [number % m for m in moduli]

def perform_operation(a_residues, b_residues, moduli, op):
    """Performs the operation on residue representations."""
    c_residues = []
    for a, b, m in zip(a_residues, b_residues, moduli):
        if op == '+':
            c = (a + b) % m
        elif op == '-':
            c = (a - b) % m
        elif op == '*':
            c = (a * b) % m
        elif op == '/':
            b_inv = mod_inverse(b, m)
            if b_inv is None:
                raise ZeroDivisionError(f"Modular inverse of {b} mod {m} does not exist.")
            c = (a * b_inv) % m
        else:
            raise ValueError("Unsupported operation.")
        c_residues.append(c)
    return c_residues

def main():
    print("Initialize the system:")
    k = int(input("Enter number of moduli (k): "))
    moduli = []
    for i in range(k):
        mi = int(input(f"Enter modulus m{i+1} (should be coprime): "))
        moduli.append(mi)

    M = 1
    for m in moduli:
        M *= m
    print(f"\nModuli: {moduli}")
    print(f"Product of moduli M = {M}\n")

    while True:
        print("\nMenu:")
        print("1. Add A + B")
        print("2. Subtract A - B")
        print("3. Multiply A * B")
        print("4. Divide A / B (modular division)")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == '5':
            print("Exiting program.")
            break

        A = int(input("Enter first large number A: "))
        B = int(input("Enter second large number B: "))

        a_residues = to_residue_repr(A, moduli)
        b_residues = to_residue_repr(B, moduli)

        print(f"A in residues: {a_residues}")
        print(f"B in residues: {b_residues}")

        try:
            if choice == '1':
                op = '+'
            elif choice == '2':
                op = '-'
            elif choice == '3':
                op = '*'
            elif choice == '4':
                op = '/'
            else:
                print("Invalid choice.")
                continue

            c_residues = perform_operation(a_residues, b_residues, moduli, op)
            print(f"Result residues (c1, ..., ck): {c_residues}")

            C = crt_reconstruct(c_residues, moduli)
            print(f"Final Result C = A {op} B mod M = {C}")

        except Exception as e:
            print(f"Error: {e}")

if __name__ == "__main__":
    main()
```
