**Problem Definition** â€“ The greatest common divisor GCD(a, b) of two natural numbers a and b is the greatest number
that divides both a and b. This number plays an important role in the number theory and public cryptography. The
algorithm is based on the following fact (assume a â‰¥ b > 0): a = q â‹… b + r; 0 â‰¤ r < b. Here q is the quotient, i.e. âŒŠð‘Ž/ð‘âŒ‹,
and r is the remainder r = a mod b. There are three algorithms to find GCD and some additional information.

**A) Basic Euclidean algorithm [Tabular Method]** for GCD â€“ This algorithm has three columns: q, r1, r2 and r. The first
column (q) is for quotient (i.e. âŒŠð‘Ÿ1/ð‘Ÿ2âŒ‹). The next two columns are r1 and r2 represents two integer numbers in
which r1 is largest number and r2 is second integer number. The last column r is remainder of division a/b (i.e. r1
mod r2). This tabular method is iterated by shifting r2 and r to r1 and r2 respectively. This process is repeated until
no further division r1/r2 is not possible. The last updated r1 represents GCD of two input numbers.

**B) Extended Euclidean algorithm [Tabular Method]** for GCD and (s,t) â€“ This algorithm has additional six columns to
Basic Euclidean algorithm. The additional six columns are s1, s2, s, t1, t2 and t. The initial values of s1 and s2 are 1
and 0 respectively. The initial values of t1 and t2 are 0 and 1 respectively. To compute q, r, s and t, this algorithm
uses the following three equations: q = âŒŠð‘Ÿ1/ð‘Ÿ2âŒ‹, r= r1 mod r2, s=s1-s2*q and t=t1-t2*q. The tabular Extended
Euclidean algorithm is iterated by performing six shifting (i.e. r2, r, s2, s, t2 and t to r1, r2, s1, s2, t1 and t2
respectively) and computing q, r, s and t in each iteration. This process is repeated until no further division r1/r2 is
not possible. The last updated r1 s1, and t1 represents GCD, s and t respectively.

**C) Euclidean algorithm [Tabular Method]** for GCD and MI â€“ This algorithm uses only 7 columns of Extended
Euclidean algorithm which are q, r1, r2, r, t1, t2 and t. This algorithm is also iterated by performing four shifts and
computation of q, r and t in each iteration. This process is repeated until no further division r1/r2 is not possible. The
last updated r1 and t1 represents GCD and MI respectively.

```python
inputs = [[4569, 8390],[4,16],[9034,4524],[39233,23499],[4573464,405703470]]
```

```python
def basicEuclidian(r1,r2):
    print(f"{'q':>3} {'r1':>5} {'r2':>5} {'r':>5}")
    while r2 != 0:
        q = r1 // r2
        r = r1 % r2
        print(f"{q:>3} {r1:>5} {r2:>5} {r:>5}")
        r1, r2 = r2, r
    print(f"\nGCD is {r1}")
    return r1
```

```python
for a,b in inputs:
  print(f"Input: {a} {b}")
  basicEuclidian(a,b)
  print()
```

```python
def extendedEuclidian(a,b):
    s1, s2 = 1, 0
    t1, t2 = 0, 1

    # Header
    print(f"{'q':>3} {'a':>5} {'b':>5} {'r':>5} {'s1':>5} {'s2':>5} {'s':>5} {'t1':>5} {'t2':>5} {'t':>5}")

    while b != 0:
        q = a // b
        r = a % b
        s = s1 - q * s2
        t = t1 - q * t2

        # Print current row
        print(f"{q:>3} {a:>5} {b:>5} {r:>5} {s1:>5} {s2:>5} {s:>5} {t1:>5} {t2:>5} {t:>5}")

        # Shift values for next iteration
        a, b = b, r
        s1, s2 = s2, s
        t1, t2 = t2, t

    # Final result
    print(f"\nGCD is {a}, s = {s1}, t = {t1}")
    return a, s1, t1
```

```python
for a,b in inputs:
  print(f"Input: {a} {b}")
  extendedEuclidian(a,b)
  print()
```

```python
def tabularEucledian(r1,r2):
    t1, t2 = 0, 1
    original_modulus = r1
    # Header
    print(f"{'q':>3} {'r1':>5} {'r2':>5} {'r':>5} {'t1':>5} {'t2':>5} {'t':>5}")

    while r2 != 0:
        q = r1 // r2
        r = r1 % r2
        t = t1 - q * t2

        # Print current row
        print(f"{q:>3} {r1:>5} {r2:>5} {r:>5} {t1:>5} {t2:>5} {t:>5}")

        # Shift for next iteration
        r1, r2 = r2, r
        t1, t2 = t2, t

    # Final output
    gcd = r1
    if gcd == 1:
        mi = t1 % original_modulus  # Ensure result is positive
        print(f"\nGCD is {gcd}, Modular Inverse = {mi}")
        return gcd, mi
    else:
        print(f"\nGCD is {gcd}, Modular Inverse does not exist.")
        return gcd, None
```

```python
for a,b in inputs:
  print(f"Input: {a} {b}")
  tabularEucledian(a,b)
  print()
```
